## The Game of Life
-----------------------------

The **Game of Life** is a game invented by the mathematician [John Conway](https://en.wikipedia.org/wiki/John_Horton_Conway). It consists of an infinite two-dimensional grid. Each square in the grid can either be 1 or 0 (This can be interpreted as 1 = **_Alive_** and 0 = **_Dead_**). The eight squares surrounding each square are considered its **_neighbours_**.

This is an iterative, self playing game -- the game will proceed without any user intervention after the initial arrangement of 1's and 0's is given. Each square is declared ***Dead*** (= 0) or ***Alive*** (= 1) according to the following rules:

* An ***_Alive_*** square will become ***_Dead_*** in the next time step if less than two of its neighbours in the current time step are ***_Alive_*** (underpopulation).
* An ***Alive*** square will become ***Dead*** in the next time step if more than three of its neighbours in the current time step are ***Alive*** (overpopulation).
* An ***Alive*** square with two or three ***Alive** neighbours will remain ***Alive*** in the next time step.
* A ***Dead*** square will become ***Alive*** if exactly three neighbours in the current time step are ***Alive***.
